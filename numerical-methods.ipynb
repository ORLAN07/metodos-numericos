```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import math as math
```

```python
plt.style.use('seaborn')

%matplotlib notebook
#%matplotlib inline
```

```python
def chart(f, x_i, x_f, num = 10000):
    x = np.linspace(x_i, x_f, num)
    math.fig, math.ax = plt.subplots()
    math.ax.plot(x, f(x), color='b')
    xmin, xmax = math.ax.get_xlim()
    ymin, ymax = math.ax.get_ylim()
    math.ax.grid(True, linestyle='-')
    math.ax.annotate("", xy=(xmax, 0), xytext=(xmin, 0), arrowprops=dict(color='gray', width=1.5, headwidth=8, headlength=10))
    math.ax.annotate("", xy=(0, ymax), xytext=(0, ymin), arrowprops=dict(color='gray', width=1.5, headwidth=8, headlength=10))
    plt.show()
```

```python

    
def f(x): 
    return (x + np.log(x))

chart(f, 0.01, 1.4)
    
```

```python
def g(x):
    math.e**x - np.sin(x)-3*x**2
    
chart(g, -1.5, 1.5)
```
