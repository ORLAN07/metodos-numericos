```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def graficar(f, x_i, x_f, num = 1000):
    x = np.linspace(x_i, x_f, num)
    fig, ax = plt.subplots(figsize=(15,5))
    ax.plot(x, f(x))
    xmin, xmax = ax.get_xlim()
    ymin, ymax = ax.get_ylim()
    ax.annotate("", xy=(xmax, 0), xytext=(xmin, 0),arrowprops=dict(color='gray', width=1.5, headwidth=8, headlength=10))
    ax.annotate("", xy=(0, ymax), xytext=(0, ymin),arrowprops=dict(color='gray', width=1.5, headwidth=8, headlength=10))
    plt.show()
    return None

graficar(f, 0.01, 1.4)

def bisection_method(f, a, b, tol=10**-5, N=500): 
    i = 1
    if f(a)*f(b) >= 0:
        print('El intervalo no funciona')
        return None
    error_abs = abs(b - a)
    while i <= N and error_abs >= tol:
        c = (a + b) / 2
        if f(c) == 0:
            print('solucion encontrada:' + c)
            return c
        
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
        error_abs = abs(b - a)
        if error_abs < tol:
            print('Solución encontrada')
            return c
        i += 1
    print('Solucion no encontrada')
    return None

def f(x): 
    return x + np.log(x)

bisection_method(f, 0.2, 1.4)
```
